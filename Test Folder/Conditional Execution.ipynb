## If Else statement

### Check wheter a person is eligible to vote

```python
# Take age of person as interger input
age = int(input("Please enter age in years : "))
# Apply if else code 
if age >= 18:
    print(f"At age : {age} years, You are Eligible to vote")
else:
    print(f"At age : {age} years, You are NOT eligible to vote")
    print(f"Still {18 - age} years required to become eligible to vote")
```

```python
# Take input from user
n = int(input("Please enter a positive number : "))

# Verify if the number is positive
assert n > 0 , "Given number should be positive"

# Apply if else loop
if n%3 == 0:
    print("Given number is DIVISIBLE by 3")
else:
    print("Given number is NOT Divisible by 3")
```

### Multiple conditions to check - if elif else

### Check wheter a number is Divisible by 3 and 7 in detail

```python
# Take the number as input
n = int(input("Please enter a positive number : "))

# Apply if elif else
if (n%3 == 0) and (n%7 == 0):
    print("The Given number is DIVISIBLE by BOTH 3 AND 7")
elif n%3 == 0:
    print("The Given number is DIVISIBLE ONLY By 3")
elif n%7 == 0:
    print("The Given number is DIVISIBLE ONLY By 7")
else:
    print("The Given number is NOT Divisible by BOTH 3 and 7")
```

### You clothing store and you want to provide a discount based on total bill
1. Upto 2000 INR - No Discount
2. 2000 to 3000 - 5% Discount
3. 3000 to 4000 - 7.5% Discount
4. 4000 and above - 10% Discount

```python
# Take amount before discount as input from user
amt = float(input("Please enter total amount : "))

# Check wether amount provided is postive and non zero
assert amt > 0, "Amount cannot be negative or zero"

# Apply if else loop
if amt < 2000:
    disc = 0
    print("No Discount upto 2000 INR")
elif amt >= 2000 and amt < 3000:
    disc = (5/100)*amt
    print("5% Discount for purchase between 2000 and 3000")
elif amt >= 3000 and amt < 4000:
    disc = (7.5/100)*amt
    print("7.5% Discount for purchase between 3000 and 4000")
else:
    disc = (10/100)*amt
    print("10% Discount for purchase above 4000")

# Calculate final billing amount
final_amt = amt - disc

# Print the Billing details
print(f"Amount Before Discount : {amt:.2f} INR")
print(f"Discount applied : {disc:.2f} INR")
print(f"Amount After Discount : {final_amt:.2f} INR")
```
